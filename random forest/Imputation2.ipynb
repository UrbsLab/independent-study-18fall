## k-Nearest Neighbor (kNN)
For k-Nearest Neighbor imputation, the missing values are based on a kNN algorithm. These values are obtained by using similarity-based methods that rely on distance metrics. KNN can be used for data that are continuous, discrete, ordinal and categorical which makes it particularly useful for dealing with all kind of missing data. The main disadvantage of using kNN imputation is that it becomes time-consuming when analyzing large datasets because it searches for similar instances through all the dataset.

## Multiple Imputation by Chained Equations(MICE)
Multiple imputation is a process where the missing values are filled multiple times to create “complete” datasets. Multiple imputation has a lot of advantages over traditional single imputation methods. Multiple Imputation by Chained Equations (MICE) is an imputation method that works with the assumption that the missing data are Missing at Random (MAR). Recall that for MAR, the nature of the missing data is related to the observed data but not the missing data. The MICE algorithm works by running multiple regression models and each missing value is modeled conditionally depending on the observed (non-missing) values.

# Part 4: Model Training and Result Evaluation

### K-fold Cross-Validation

```python
from sklearn.cross_validation import KFold
```

```python
# This function does 10-fold. It saves the result at each time as different parts of y_pred. 
# In the end, it returns the y_pred as the result of all the 10-fold.
def run_cv(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=10,shuffle=True) # Total number of elements；Number of folds， default=3；Whether to shuffle the data before splitting into batches
    y_pred = y.copy()
    clf = clf_class(**kwargs)
    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred
```

### Run Random Forest Model and Calculate Accuracy

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# This function calculates accuracy
def accuracy(y_true,y_pred):
    return np.mean(y_true == y_pred) # NumPy interpretes True and False as 1. and 0.
```

```python
# This function take a dataframe and return X and y for model training
def split_dataset(dataframe):
    X = dataframe.iloc[:, :-1].values
    y = dataframe.iloc[:, -1].values
    return X, y
```

```python
X_10_knn, y_10_knn = split_dataset(data_10_knn)
X_20_knn, y_20_knn = split_dataset(data_20_knn)
X_50_knn, y_50_knn = split_dataset(data_50_knn)

X_10_mice, y_10_mice = split_dataset(data_10_mice)
X_20_mice, y_20_mice = split_dataset(data_20_mice)
X_50_mice, y_50_mice = split_dataset(data_50_mice)
```

```python
RF_CV_result_10_knn = run_cv(X_10_knn, y_10_knn,RandomForestClassifier)
RF_CV_result_20_knn = run_cv(X_20_knn, y_20_knn,RandomForestClassifier)
RF_CV_result_50_knn = run_cv(X_50_knn, y_50_knn,RandomForestClassifier)

RF_CV_result_10_mice = run_cv(X_10_mice, y_10_mice,RandomForestClassifier)
RF_CV_result_20_mice = run_cv(X_20_mice, y_20_mice,RandomForestClassifier)
RF_CV_result_50_mice = run_cv(X_50_mice, y_50_mice,RandomForestClassifier)
```

```python
print ("Random forest accuracy using knn imputation on the dataset with 10% missing data:"+str(accuracy(y_10_knn, RF_CV_result_10_knn)))
print ("Random forest accuracy using knn imputation on the dataset with 20% missing data:"+str(accuracy(y_20_knn, RF_CV_result_20_knn)))
print ("Random forest accuracy using knn imputation on the dataset with 50% missing data:"+str(accuracy(y_50_knn, RF_CV_result_50_knn)) + "\n")

print ("Random forest accuracy using MICE imputation on the dataset with 10% missing data:"+str(accuracy(y_10_mice, RF_CV_result_10_mice)))
print ("Random forest accuracy using MICE imputation on the dataset with 20% missing data:"+str(accuracy(y_20_mice, RF_CV_result_20_mice)))
print ("Random forest accuracy using MICE imputation on the dataset with 50% missing data:"+str(accuracy(y_50_mice, RF_CV_result_50_mice)) + "\n")
```
