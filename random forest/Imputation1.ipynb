# Part 2: Create random missing values

```python
import random
data10 = raw_data.copy()
data20 = raw_data.copy()
data50 = raw_data.copy()
ix = [(row, col) for row in range(1600) for col in range(20)]
 
for row, col in random.sample(ix, int(round(.1*1600*20))): #randomly select 10% of values to be None
    data10.iloc[row, col] = None
    
for row, col in random.sample(ix, int(round(.2*1600*20))): #randomly select 20% of values to be None
    data20.iloc[row, col] = None

for row, col in random.sample(ix, int(round(.5*1600*20))): #randomly select 50% of values to be None
    data50.iloc[row, col] = None
```

# Part 3: Imputation

## Mean  
The simplest way is to fill the missingd data with mean of the column. <br>
It has the advantage of keeping the same mean and the same sample size, but has many disadvantages as well. Mean imputation does not preserve the relationships among variables.It will bias your standard error. The standard error of that variable will be vastly underestimated–and this underestimation gets bigger the more missing data there are. Too-small standard errors lead to too-small p-values. Now you’re making Type I errors without realizing it.

```python
data10_mean = data10.copy()
data20_mean = data20.copy()
data50_mean = data50.copy()
```

```python
data10_mean = data10_mean.fillna(data10.mean())
```

```python
data20_mean = data20_mean.fillna(data20.mean())
```

```python
data50_mean = data50_mean.fillna(data50.mean())
```

## Median
If datasets have great outliers,results in the mean of those datasets are biased by high magnitude outliers. in this situation, replacing missing data with median might be a good choice.

```python
data10_median = data10.copy()
data20_median = data20.copy()
data50_median = data50.copy()
```

```python
data10_median = data10_median.fillna(data10.median())
```

```python
data10_median.head()
```

```python
data20_median = data20_median.fillna(data20.median())
```

```python
data20_median
```

```python
data50_median = data50_median.fillna(data50.median())
```

```python
data50_median.head()
```

## Mode
Mode imputation replaces missing values of a categorical variable by the mode of non-missing cases of that variable.

```python
data10_mode = data10.copy()
data20_mode = data20.copy()
data50_mode = data50.copy()
```

```python
data10_mode = data10_mode.fillna(data10.mode().iloc[0])#Note that there could be multiple values returned for the selected axis (when more than one item share the maximum frequency)
```

```python
data10_mode.head()
```

```python
data20_mode = data20_mode.fillna(data20.mode().iloc[0])
```

```python
data20_mode.head()
```

```python
data50_mode = data50_mode.fillna(data50.mode().iloc[0])
```

# Part 4: Model Training and Result Evaluation

### K-fold Cross-Validation

```python
from sklearn.cross_validation import KFold
```

```python
# This function does 10-fold. It saves the result at each time as different parts of y_pred. 
# In the end, it returns the y_pred as the result of all the 10-fold.
def run_cv(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=10,shuffle=True) # Total number of elements；Number of folds， default=3；Whether to shuffle the data before splitting into batches
    y_pred = y.copy()
    clf = clf_class(**kwargs)
    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred
```

### Run Random Forest Model and Calculate Accuracy

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# This function calculates accuracy
def accuracy(y_true,y_pred):
    return np.mean(y_true == y_pred) # NumPy interpretes True and False as 1. and 0.
```

```python
# This function take a dataframe and return X and y for model training
def split_dataset(dataframe):
    X = dataframe.iloc[:, :-1].values
    y = dataframe.iloc[:, -1].values
    return X, y
```

```python
X_10_mean, y_10_mean = split_dataset(data10_mean)
X_20_mean, y_20_mean = split_dataset(data20_mean)
X_50_mean, y_50_mean = split_dataset(data50_mean)

X_10_median, y_10_median = split_dataset(data10_median)
X_20_median, y_20_median = split_dataset(data20_median)
X_50_median, y_50_median = split_dataset(data50_median)

X_10_mode, y_10_mode = split_dataset(data10_mode)
X_20_mode, y_20_mode = split_dataset(data20_mode)
X_50_mode, y_50_mode = split_dataset(data50_mode)
```

```python
RF_CV_result_10_mean = run_cv(X_10_mean, y_10_mean,RandomForestClassifier)
RF_CV_result_20_mean = run_cv(X_20_mean, y_20_mean,RandomForestClassifier)
RF_CV_result_50_mean = run_cv(X_50_mean, y_50_mean,RandomForestClassifier)

RF_CV_result_10_median = run_cv(X_10_median, y_10_median,RandomForestClassifier)
RF_CV_result_20_median = run_cv(X_20_median, y_20_median,RandomForestClassifier)
RF_CV_result_50_median = run_cv(X_50_median, y_50_median,RandomForestClassifier)

RF_CV_result_10_mode = run_cv(X_10_mode, y_10_mode,RandomForestClassifier)
RF_CV_result_20_mode = run_cv(X_20_mode, y_20_mode,RandomForestClassifier)
RF_CV_result_50_mode = run_cv(X_50_mode, y_50_mode,RandomForestClassifier)
```

```python
print ("Random forest accuracy using mean imputation on the dataset with 10% missing data:"+str(accuracy(y_10_mean, RF_CV_result_10_mean)))
print ("Random forest accuracy using mean imputation on the dataset with 20% missing data:"+str(accuracy(y_20_mean, RF_CV_result_20_mean)))
print ("Random forest accuracy using mean imputation on the dataset with 50% missing data:"+str(accuracy(y_50_mean, RF_CV_result_50_mean)) + "\n")

print ("Random forest accuracy using median imputation on the dataset with 10% missing data:"+str(accuracy(y_10_median, RF_CV_result_10_median)))
print ("Random forest accuracy using median imputation on the dataset with 20% missing data:"+str(accuracy(y_20_median, RF_CV_result_20_median)))
print ("Random forest accuracy using median imputation on the dataset with 50% missing data:"+str(accuracy(y_50_median, RF_CV_result_50_median)) + "\n")

print ("Random forest accuracy using mode imputation on the dataset with 10% missing data:"+str(accuracy(y_10_mode, RF_CV_result_10_mode)))
print ("Random forest accuracy using mode imputation on the dataset with 20% missing data:"+str(accuracy(y_20_mode, RF_CV_result_20_mode)))
print ("Random forest accuracy using mode imputation on the dataset with 50% missing data:"+str(accuracy(y_50_mode, RF_CV_result_50_mode)))
```
